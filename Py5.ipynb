### Careful with copying!
Array slices are views on the underlying objects. If we want to create new arrays we need to explicitly .copy() them

```python
print(a2)
b = a2[1]
print(b)
b[:] = 2 #Broadcasting => utiliza el mismo n√∫mero para todos los 
#elementos del array, lo difunde
print(b)
```

```python
c = a2[1].copy()
c[:] = 10
print(a2)
print(c)
```

The base pandas abstraction. You can thing of it as the love child of a numpy array and a dictionary.

We can create Series from dictionaries:

```python
sdata = {'B':3e6,'M':6e6,'P':1.2e5,'P':7e5}
s3 = pd.Series(sdata)
print(s3)
```

```python
increase = {'B':2e5,'M':4e5,'Z':-2e4}
s4 = pd.Series(increase)
print(s4)
```

And here is where the magic happens: numpy arrays only identify their contents by position. In contrast, pandas knows their "name" and will align them based on their indexes:

```python
print(s3 + s4)
print(s3.values + s4.values)
```

```python
s3.name = 'population_2000'
s3.index.name = 'province'
print(s3)
```

### Dropping entries from an axis

```python
s5 = pd.Series(np.arange(5),list('jduvk'))
print(s5)
```

```python
#Descartar elementos
print(s5.drop(['d','k'])) #No borra la serie, simplemente muestra una nueva 
#serie sin la selecci√≥n
#De hecho s5 no cambia
print(s5)
```

By default, `drop()` doesn't modify the original Series- it creates a copy. We can change that with the argument `inplace`.

```python
df2.drop('c')
```

```python
df2.drop('2nd_language',axis=1) #axis = 1 indica que es orientaci√≥n 
#vertical
```

```python
df3 = df2.copy()
df3
```

```python
#Modificar DataFrame original pero s√≥lo lo puedes ejecutar una √∫nica vez
df3.drop('capital',axis=1,inplace=True)
df3
```

###  Indexing, selection, and filtering
The key here is that we can build boolean Series that we can use to index the original Series or DataFrame. Those booleans can be combined with bitwise boolean operators (&, |, ~) to get filters that are as complex as we need. 

```python
print(s3)
```

```python
print(s3[['V','M']]) #Aqu√≠ la serie se comporta como un diccionario
```

```python
print(s3[2:]) #Aqu√≠ se comporta como un array
```

```python
print(s3['P':'V']) #Se puede comportar como los dos
```

```python
print(s3 > 1e06) #Y esto lo podemos utilizar para indexar la propia serie
```

```python
print(s3[s3 > 1e06])
```

```python
#devuelveme los elemetos de df3 que tengan los a√±os posteriores a 1950
df3[df3['year'] > 1950]
```

```python
#devuelveme los anteriores a 1900 y que tengan una deuda mayor que 1
df3[(df3['year']>1900)&(df3['debt']>1)]
```

```python
#tambi√©n se puede hacer indexando sobre el DataFrame original y el volver a indexar el DataFrame resultante.
df3[(df3['year']>1900)][(df3['debt']>1)]
```

```python
recent= df3['year']>1900
indebted=df3['debt']>1
df3[recent&indebted]
```
